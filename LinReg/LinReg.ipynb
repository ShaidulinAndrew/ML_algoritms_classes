```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

__Загрузка данных__

```python
boston_df = pd.read_csv('./boston.csv')
boston_df.head()
```

```python
# посмотрим на тип переменных с которыми нам предстоит работать
boston_df.info()
```

```python
# посмотрим на основные статистические показатели (summery statistics)
# с помощью метода .describe()
boston_df.describe().round(2)
```

__Предварительная обработка данных__

```python
# проверим есть ли пропущенные значения в наборе данных
boston_df.isnull().sum()
```

__Исследовательский анализ данных__

```python
# посчитаем коэффициент корреляции для всего dataframe и округлим значения
# получим корреляционную матрицу
corr_matrix = boston_df.corr().round(2)
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
```

```python
# построим диаграмму рассеяния между целевой переменной MEDV и LSTAT и RM
# сначала подготовим данные для визуализации

x1 = boston_df['LSTAT']
x2 = boston_df['RM']
y = boston_df['MEDV']
```

```python
# зададим размер и построим первый график
plt.figure(figsize=(7,4))
plt.scatter(x1, y)

# добавим подписи
plt.xlabel('Процент населения с низким социальным статусом', fontsize=8)
plt.ylabel('Медианная цена недвижимости, тыс. долларов', fontsize=8)
plt.title('Социальный статус населения и цены на жилье', fontsize=10)
```

```python
# зададим размер и построим первый график
plt.figure(figsize=(7,4))
plt.scatter(x2, y)

# добавим подписи
plt.xlabel('Количество комнат', fontsize=8)
plt.ylabel('Медианная цена недвижимости, тыс. долларов', fontsize=8)
plt.title('Количество комнат и цены на жилье', fontsize=10)
```

__Отбор и выделение признаков__

```python
# Отберем признаки с наиболее высокой корреляцией с целевой переменной
# и поместим их в переменную X

X = boston_df[['RM', 'LSTAT', 'PTRATIO', 'TAX', 'INDUS']]

# целевую переменную поместим в переменную y

y = boston_df['MEDV']
```

```python
# разбиваем данные на обучающую и тестовую выборки

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# посмотрим на размерность обучающей и тестовой выборки
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

__Обучение и оценка качества модели__

```python
from sklearn.linear_model import LinearRegression

# создадим объект этого класса и запишем в переменную model
model = LinearRegression()

# обучим нашу модель
model.fit(X_train, y_train)
```

```python
# предскажем стоимость на жилье
y_pred = model.predict(X_test)

# выведем первые 5 значений предсказаний
print(y_pred[:5])
```

```python
# импортируем модуль с метриками
from sklearn import metrics

# выведем корень среднеквадратической ошибки 
# сравним тестовые и прогнозные значения цен на жилье
print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# посмотрим на еще одну метрику, R2 (коэффициент детерминации) 
# он показывает какую часть изменчивости целевой переменной мы можем объяснить с помощью модели (чем больше тем лучше)

print('R2:', np.round(metrics.r2_score(y_test, y_pred), 2))
```
